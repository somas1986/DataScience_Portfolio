```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#reading csv
df_vaccines=pd.read_csv('vaccinations.csv')
```

```python
df_vaccines.head(100)
```

# 1. Drop Columns

Dropping columns which are not required is one of the important data transformation as it reduces the data size for processing. I have removed daily_vaccinations_raw, daily_vaccinations, total_vaccinations_per_hundred, people_vaccinated_per_hundred, people_fully_vaccinated_per_hundred, total_boosters_per_hundred, daily_vaccinations_per_million, daily_people_vaccinated,daily_people_vaccinated_per_hundred columnns as they are not required and doesn't provide any value for this project. 

```python
df_vaccines.drop(df_vaccines.iloc[:, 7:], inplace=True, axis=1)
```

```python
df_vaccines.head()
```

# 2. Rename Columns

I have renmaed couple of columns location and iso_code to country_name and country_code as the earlier names were not clearly providing info as what they represent. Location can be city, state, country or latitude and longitude combination.

```python
columns_list=['country_name','country_code','date','total_vaccinations','people_vaccinated','people_fully_vaccinated','total_boosters_administered']
```

```python
# replacing the column names
df_vaccines.columns=columns_list
df_vaccines.head()
```

# 3. Missing Values

Many important columns has null values or missing values as null data doesnt yeild any value, I am droping records with null values

```python
df_vaccines.isnull().sum()
```

```python
df_vaccines.shape
```

```python
df_vaccines = df_vaccines.dropna()
```

```python
df_vaccines.shape
```

```python
df_vaccines.head(10)
```

# 4. Convert the data types

I have convertes data types of total_vaccinations, people_vaccinated, people_fully_vaccinated and total_boosters_administered to int as float doesnt have any sense to these columns. For example, There can not be 1.5 total vaccinations in a country i.e., 1.5 people getting vaccinated is wrong. 

```python
df_vaccines["total_vaccinations"] = df_vaccines["total_vaccinations"].astype(int)
df_vaccines["people_vaccinated"] = df_vaccines["people_vaccinated"].astype(int)
df_vaccines["people_fully_vaccinated"] = df_vaccines["people_fully_vaccinated"].astype(int)
df_vaccines["total_boosters_administered"] = df_vaccines["total_boosters_administered"].astype(int)
```

```python
df_vaccines.head()
```

# Unique Country name count

```python
df_vaccines['country_name'].nunique()
```

# 5. Sort dataframe

Sorting the data frame to organize the data in descending order to keep the latest data at the top and in the next step we can drop other duplicate values peer country

```python
df_vaccines.sort_values(by=['country_name','country_code','date','total_vaccinations'],ascending=False)
```

# 6. Drop duplicates

Dropping duplicate values per country and retaining just one record per country. Count of unique country name count of 205 is matching with count of records after droppinf duplicate records. So, it shows we have achieved the target.

```python
#removing duplicates
df_vaccines.drop_duplicates(subset='country_name',keep='first',inplace=True)
```

```python
df_vaccines
```

# 7. Fix casing for columns with string data type

Converted country_name and country_code to upper case as when we combine this data with other data sources based on these columns, consistent casing will help us in combining data easily.

```python
df_vaccines['country_name'] = df_vaccines['country_name'].str.upper()
df_vaccines['country_code'] = df_vaccines['country_code'].str.upper()
```
