```python
#importing libraries
import numpy as np
import pandas as pd
from scipy import stats
```

```python
# for handling http requests
import requests 
```

```python
# for html parsing and web scraping
from bs4 import BeautifulSoup 
import bs4 as bs
import re
import urllib
from urllib.request import Request, urlopen 
```

```python
#webscraping
url = 'https://en.wikipedia.org/wiki/COVID-19_testing'
```

```python
#read from url
html = urlopen(url) 
soup = BeautifulSoup(html, 'html.parser')
```

```python
#read wikitables
tables = soup.find_all(id='wikitable')
covid_table = soup.findAll("table", { 'class' : 'wikitable' })
```

```python
#As our data is in second table, read data from second table and load to Dataframe
df_table = pd.read_html(str(covid_table))

df_covid = df_table[1]
df_covid
```

# 1. Replace headers

Replaced headers to format which is easy to use in porgramming as the current headers some unwanted wordings which might even confuse the developer.

```python
df_covid.rename(columns={'Country or region': 'country', 'Date[a]': 'date','Tested': 'tested', 'Units[b]': 'units','Confirmed(cases)':'confirmed_cases','Confirmed /tested,%':'percentage_confirmed','Tested /population,%':'percentage_tested','Confirmed /population,%':'percentage_confirmed_population'}, inplace=True)
```

```python
df_covid
```

# 2. Drop Columns

As we have no value in having columns units and Ref, I will be dropping these columns. These columns doesnt add any value for your analysis.

```python
df_covid = df_covid.drop(['units','Ref.'], axis=1)
```

```python
df_covid
```

# 3. Drop Rows

As the last rows in the table of Wikipedia has the legend, we do not need that row for our analysis. I am dropping this last row.

```python
df_covid = df_covid[:-1]
```

```python
df_covid
```

# 4.  Fix casing for columns with string data type

Converting the country name to upper case so that when we work 2 other datasets, it will be easy with merging the datasets based on country name

```python
df_covid['country'] = df_covid['country'].str.upper()
```

```python
df_covid
```

# 5. Change data types of the columns

Changing data type of all the columns as we need to use this data to apply transformations. For now, this will help changing date format to YYYY-MM-dd.

```python
df_covid = df_covid.astype({'date': 'datetime64', 'tested': 'int', 'confirmed_cases': 'int', 'percentage_confirmed': 'float', 'percentage_tested': 'float', 'percentage_confirmed_population': 'float'})
```

```python
df_covid
```

# 6. Identify outliers and bad data

Identifying the outliers is very important as this might lead to wrong results. Hence, applied data cleaning of identifying outliers. As this a public site, we can not trust the data quality.

```python
df_covid_new = df_covid[
    (np.abs(stats.zscore(df_covid.select_dtypes(include=np.number))) > 3).all(axis=1)
]
```

```python
df_covid
```

# 7. Drop duplicates

Manually, with lesser data we might be able to identify the duplicates. But when we run this program regularly and the data gets updated and a case where we can find duplicates on nth day of program run, we can get bad results. So, I would like to handle it programatically.

```python
df_covid = df_covid.drop_duplicates()
```

# Ethical Implications of Data Wrangling:

I have used COVID testings data from Wikipedia. Wikipedia is a public site and the data posted is visible to the public. Ethical implications of data wrangling of Wikipedia data:
    1. As we access data from the URL directly, this increases number of hits on the server which will have performance issues on the website.
    2. As public has access to update the data on Wikipedia, we can not consider this as trusted source and the data might not be upto date. This means the data being used might not be accurate.
    3. As there are multiple exceptions stated in the legend of the Wikitable, some of the countries data excluded few people or few areas of the country. This means numbers provided for few countries is not accurate.
    4. As we are performing some clean-up and transformations, we might change some data and data being used in our analysis might be slightly different from source data.
    
   As this is a public website, we shouldn't have any legal complications. However, the data wrangling from this source has implications of accuracy and performance issues on the website.
